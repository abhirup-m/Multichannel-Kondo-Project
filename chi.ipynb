```python
import numpy as np
from qutip import *
from tqdm import tqdm
from multiprocessing import Pool
import matplotlib
from matplotlib import pyplot as plt
plt.rc('font', size=30)          # controls default text sizes
plt.rc('xtick', labelsize=30)    # fontsize of the tick labels
plt.rc('ytick', labelsize=30)    # fontsize of the tick labels
plt.rc('legend', fontsize=30)    # legend fontsize
plt.rc('lines', markersize = 10)

matplotlib.rcParams['text.usetex'] = True
plt.rcParams["figure.figsize"]= 15, 10
plt.rcParams['figure.dpi'] = 70
matplotlib.rcParams['lines.linewidth'] = 4
plt.rcParams['axes.grid'] = False
plt.rcParams['axes.grid.which'] = "both"
props = dict(boxstyle='round', alpha=0.01)
```

```python
def ham(J,K,h1,h2):
    sz = sigmaz()/2
    sx = sigmax()/2
    sy = sigmay()/2
    H = 0
    for ch in range(K):
        H += J*tensor([sx] + [identity(2)]*ch + [sx] + [identity(2)]*(K-ch-1))
        H += J*tensor([sy] + [identity(2)]*ch + [sy] + [identity(2)]*(K-ch-1))
        H += J*tensor([sz] + [identity(2)]*ch + [sz] + [identity(2)]*(K-ch-1))
        H += h2*tensor([identity(2)]*(1+ch) + [sz] + [identity(2)]*(K-ch-1))
    H += h1*tensor([sz] + [identity(2)]*K)
    return H
```

```python
def get_Z(J,K,h1,h2,T):
    beta = 1/T
    H = ham(J,K,h1,h2)
    E = H.eigenenergies()
    Z = np.sum(np.exp(-beta*E))
    return Z

def get_chi_numerically(J,K,T_range):
    h2 = 0
    h1 = 0
    deltah = 0.0001
    chi = []
    for T in T_range:
        Z = get_Z(J,K,h1,h2,T)
        Z_plus_deltaZ = get_Z(J,K,h1+deltah,h2,T)
        Z_plus_2deltaZ = get_Z(J,K,h1+2*deltah,h2,T)
        dZdh = (Z_plus_deltaZ - Z)/deltah
        d2zdh2 = (Z_plus_2deltaZ + Z - 2*Z_plus_deltaZ)/(deltah**2)
        chi.append(T * (d2zdh2/Z - (dZdh/Z)**2))
    return chi
```

```python
def get_chi_exact(J,K,T_range):
    M_max = K/2
    M_min = 0 if K%2 == 0 else 1/2
    M_range = np.arange(M_min, M_max+0.1)
    chi = []
    for T in T_range:
        beta = 1/T
        Z = 0
        Zp = 0
        Zpp = 0
        for M in M_range:
            m_range = np.arange(-M, M-1+0.1)
            t1 = 1 + (m_range+1/2)**2
            t2 = np.sqrt(t1)
            Z += 2*np.exp(-beta*J*M/2)
            Z += 2*np.exp(beta*J/4)*np.sum(np.cosh(beta*J*t2/2))
            Zp += (np.exp(beta*J/4)*beta)*np.sum((m_range+0.5)*np.sinh(beta*J*t2/2)/t2)
            Zpp += np.sum(np.sinh(beta*J*t2/2)/t2)
            Zpp += np.sum(((beta*J/2)*(m_range+0.5)**2*np.cosh(beta*J*t2/2)/t2)/t2)
            Zpp -= np.sum(((m_range+0.5)**2*np.sinh(beta*J*t2/2)/t1)/t2)
            Zpp *= beta*np.exp(beta*J/4)/J
            Zpp += (beta**2 / 2) * np.sum(np.exp(-beta*J*M/2))
        chi.append(T * (Zpp/Z - (Zp/Z)**2))
    return chi

J = 1
T_range = 10**np.linspace(-2,0,50)
K = 2
J = 1/K
chi_exact = get_chi_numerically(J,K,T_range)
plt.scatter(T_range, chi_exact,marker='.')
beta = 1/T_range
chi = (beta/2 + (1/J)*np.exp(beta*J/4)/(5*(1 + np.exp(-beta*J/2)))*(16*np.sinh(beta*J*np.sqrt(5)/4)/np.sqrt(5) + beta*J*np.cosh(beta*J*np.sqrt(5)/4)))/(2 + 4*np.cosh(beta*J*np.sqrt(5)/4)*np.exp(beta*J/4)/(1+np.exp(-beta*J/2)))
plt.plot(T_range, chi)
plt.xscale("log")
```
